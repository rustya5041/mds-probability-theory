`1.1 Prove that for symmetrical PDFs median is equal to expected value.`

To prove that, we need to confirm that the point of symmetry of the PDF is equal to the median, then that the point of symmetry is equal to the $\mathbb{E}(X)$.
Let us start by proving that the point of symmetry is equal to the median.

Assume $x{_0}$ is the point of symmetry
$\implies$ it should follow that $\int_ {-\infty} ^{X{_0}} p(x)dx$ = $\int_ {x{_0}} ^\infty p(x)dx$ 

$\forall x{_0} - \Delta x$ we obtain $\int_ {-\infty} ^{X{_0}} p(x{_0} - \Delta x)dx = \int_ 0 ^\infty p(x{_0} - \Delta x)dx$  
$\forall x{_0} + \Delta x$ we obtain $\int_ {x{_0}} ^\infty p(x{_0} + \Delta x)dx = \int_ 0 ^\infty p(x{_0} - \Delta x)dx$  
$\therefore$ the point of symmetry equals to the median value of the pdf.

Now let us prove that $\mathbb{E}(X)$ = point of symmetry. And **by definition**, if $\mathbb{E}(X)$ exists (and it does, which follows from the task condition) and pdf is symmetric with respect to $X = X{_0}$, then $\mathbb{E}(X)$ = point of symmetry.  
 $\therefore$ $\mathbb{E}(X) = X{_0}$

The code below also highlights that the proof is correct.


```python
import matplotlib.pyplot as plt
import scipy.stats 
import numpy as np


def solution(x, pdf):
    med = np.median(x)
    exp_val = abs(round(np.sum(x * pdf) * (x[1] - x[0]), 5))
    print(f'Median: {med}\nExpected value: {exp_val}')
 
    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf, label='PDF')
    plt.axvline(med, ls='--', color='r', label='Median')
    plt.axvline(exp_val, ls=':', color='b', label='$EX$')
    plt.text(med, .15, f'Median: {med:.2f}', color='black')
    plt.text(exp_val, .17, f'$EX$: {exp_val:.2f}', color='black')
    plt.xlabel('x')
    plt.ylabel('PDF')
    plt.title('Visual proof for the given distribution')
    plt.legend()
    plt.show()


# checking  for a normal distribution
x = np.linspace(-15, 15, 1000)
pdf = scipy.stats.norm.pdf(x, loc=0, scale=2)

solution(x, pdf)
```

`1.2 Provide an example of PDF such that median is larger than the expected value.`

Let us consider some arbitrary the exponensial distribution. 
The pdf of the exponential distribution is given by $f(x) = \lambda e^{-\lambda x}$, where $\lambda$ is the rate parameter.  
The median of the exponential distribution is given by $m = \frac{1}{\lambda}$, where $\lambda$ is the rate parameter.

Let us assume that $\lambda = 1$ and $m = 1$.

The expected value of the exponential distribution is given by $\mathbb{E}(X) = \frac{1}{\lambda} = \frac{1}{1} = 1$

The code below also highlights that the proof is correct.


```python
x = np.linspace(0, 10, 1000)
pdf = scipy.stats.expon.pdf(x, loc=0, scale=1)

solution(x, pdf)
```
